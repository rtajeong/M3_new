# Something before we start Python
- In the **assignment** statement, **the expression on the right is evaluated first. That value is then assigned to the variable on the left.**
  - When assigning a variable in Python, you are creating a reference to the object (i.e pointer) , 메모리의 어딘가에 그 object를 만들어 저장하고 (그릇) 그 주소를 알려줌 (그릇의 장소).
  - **If you assign the variable to a new variable, you are only assigning the reference**, not creating a new object. (try the code below)
- Assignment in Python means one thing, and one thing only: The variable named on the left should now refer to the value on the right. In other words, when I said: y = x. Python doesn't read this as, “y should now refer to the variable x.” Rather, it read it as, **“y should now refer to whatever value x refers to.”**

```python
# mutation and iteration - be careful !
# do not mutate a list as you iterating over it !

def remove_dups(L1, L2):   # wrong code
    for e in L1:
        if e in L2:
            print(e)     # use for debugging
            L1.remove(e)

L1 = [1,2,3,4]
L2 = [1,2,5,6]
remove_dups(L1, L2)   # L1 is [2,3,4] not [3,4], why?
print(L1)
```

- L1 is [2,3,4] not [3,4] Why?
  - Python uses an internal counter to keep track of index it is in the loop
  - mutating changes the list length but Python doesn’t update the counter
  - loop never sees element 2

```python
# correct one
def remove_dups(L1, L2):
    L1_copy = L1[:]   # clone list first (or, Li_copy = L1.copy())
                      # note that L1_copy = L1 does not clone
    for e in L1_copy:
        if e in L2:
            print(e)
            L1.remove(e)
L1 = [1,2,3,4]
L2 = [1,2,5,6]
remove_dups(L1, L2)
print(L1)
```

```python
id(L1.copy()), id(L1), L1.copy(), L1  # different object (copied)
```

# Comprehension Examples
#### A = range(6), B = [‘a’, ‘b’ ‘c’] = ‘abc’, C = [(3,4),(7,24)]
- [x for x in A] ---> [0,1,2,3,4,5]
- [3*x for x in A] --->[0,3,6,9,12,15]
- [x for x in A if x < 2] ---> [0,1]
- [(x, x**2) for x in A if x % 2 == 0] --->[(0,0),(2,4),(4,16)]
- [(x,y) for x in A for y in B if x == 3] --->[(3,‘a’), (3, ‘b’), (3,‘c’)]
- [np.sqrt(x*x+y*y) for x,y in C] --->[5,25]
- {val:i for i, val in enumerate(B)} --->{‘a’:0, ‘b’:1, ‘c’:2}

# Decorator
- a decorator is a design pattern that allows you to add new functionality to an existing function (or class) without permanently modifying its code.
- It provides a way to "wrap" a function and alter its behavior in a controlled manner.

- The following shows that the function/method/class you're defining after a decorator is just basically passed on as an argument to the function/method immediately after the @ sign.
- https://www.programiz.com/python-programming/decorator

```python
def first(msg):
    print(msg)

first("Hello")

second = first  # everything is an object (including functions)
second("Hello")
```

```python
# Functions can be passed as arguments to another function.
def inc(x):
    return x + 1

def dec(x):
    return x - 1

def operate(func, x):
    result = func(x)
    return result
```

```python
print(operate(inc,3))
print(operate(dec,3))
```

- Functions and methods are called callable as they can be called.

```python
def make_pretty(func):
    def inner():
        print("I got decorated")
        func()
    return inner

def ordinary():
    print("I am ordinary")
```

```python
# one more example
# decorating functions with parameters
def divide1(a, b):
    return a/b
divide1(2,5)
```

```python
try:
    divide1(2,0)
except:
    print("Error: division by zero")
```

- Now, we want to make a decorator to check for this case that will cause the error.

```python
def smart_divide(func):
    def inner(a, b):
        print("I am going to divide", a, "by", b)
        if b == 0:
            print("Whoops! cannot divide by 0.")
            return

        return func(a, b)
    return inner

@smart_divide
def divide2(a, b):
    print(a/b)
```

```python
divide2(2, 5)
```

```python
divide2(2, 0)
```

```python
# functions can be decorated multiple times
def star(func):
    def inner(*args, **kwargs):
        print("*" * 30)
        func(*args, **kwargs)
        print("*" * 30)
    return inner

def percent(func):
    def inner(*args, **kwargs):
        print("%" * 30)
        func(*args, **kwargs)
        print("%" * 30)
    return inner

def printer1(msg):
    print(msg)

@star
@percent
def printer2(msg):
    print(msg)

printer1("Hello")
```

```python
printer2("Hello")
```

```python
# is equivalent to  (order matters)
def printer3(msg):
    print(msg)
printer = star(percent(printer3))
printer("Hello")
```

- one more example (한국말 설명)
  - Wrapper나 Wrapper Function을 이용해 어떤 함수들의 실행 이전이나 이후에 공통으로 선행해야하는 기능들이 실행되게 하는 기법
  - 클래스나 함수로 Decorator 설정가능

```python
# one more example
def decorator_function(original_function):
    def wrapper_function():
        print('{} 함수가 호출되기전 입니다.'.format(original_function.__name__))
        return original_function()
    return wrapper_function

@decorator_function
def display_1():
    print('display_1 함수가 실행됐습니다.')

@decorator_function
def display_2():
    print('display_2 함수가 실행됐습니다.')

display_1()
display_2()
```

```python
# let's compare execution time (Graph version is faster!)
# but, not always...

import timeit
class SequentialModel(tf.keras.Model):
  def __init__(self, **kwargs):
    super().__init__(**kwargs)
    self.flatten = tf.keras.layers.Flatten(input_shape=(28, 28))
    self.dense_1 = tf.keras.layers.Dense(128, activation="relu")
    self.dropout = tf.keras.layers.Dropout(0.2)
    self.dense_2 = tf.keras.layers.Dense(10)

  def call(self, x):      # forward path
    x = self.flatten(x)
    x = self.dense_1(x)
    x = self.dropout(x)
    x = self.dense_2(x)
    return x

input_data = tf.random.uniform([60, 28, 28])

eager_model = SequentialModel()
graph_model = tf.function(eager_model)

print("Eager time:", timeit.timeit(lambda: eager_model(input_data), number=10000)) # number: no of times to execute
print("Graph time:", timeit.timeit(lambda: graph_model(input_data), number=10000))
```

### Multiple inheritance

```python
# multiple inheritance
class Family:
    familyword = ""
    def show_family(self):
        print("This is our family:")

# Father class inherited from Family
class Father(Family):
    fathername = ""
    def show_father(self):
        print(self.fathername)

# Mother class inherited from Family
class Mother(Family):
    mothername = ""
    def show_mother(self):
        print(self.mothername)

# Son class inherited from Father and Mother classes
class Son(Father, Mother):
    sonname = ""
    def show_parent(self):
        print("Father :", self.fathername)
        print("Mother :", self.mothername)
        print("Son :", self.sonname)

s1 = Son()  # Object of Son class
s1.familyword = "truth"
s1.fathername = "Mark"
s1.mothername = "Sonia"
s1.sonname = "Michael"
```

# Exeptions and Assertions
- try and except: exceptions raised by any statement in body of *try* are
handled by the *except* statement and execution continues
with the body of the *except* statement
- else: body of this is executed when execution of associated
try body **completes with no exceptions**
- finally: body of this is **always executed** after try, else and
except clauses, even if they raised another error or
executed a break, continue or return

- assert: want to be sure that assumptions on state of
computation are as expected

# Python coding Example (2) : Squre Root of x
- simple guess and check algorithm
  - works only for x > 1
  - why? if x < 1, the squaring will always result in an even smaller number (e.g., 0.5 squared is 0.25). This means the guess will never be high enough to reach the target number, causing the algorithm to get stuck in an infinite loop of progressively smaller guesses.
  - make it greater than 1 by multiplying a big number (e.g. 100) and adjust it.
- Newton-Raphson method (xi+1 = xi - f(xi)/f'(xi))
  - start with a guess, g
  - if g*g is close enough to x, stop (g is the answer)
  - otherwise, make a new guess by averaging g and x/g
  - using the new guess, repeat process

```python
# Guess anf Check algotithm (only works for x >= 1) - poor algorithm
import numpy as np
epsilon = 0.01

def gc_sqrt(x):
    if x < 0.0:
        return (np.nan)
    iter = 0
    left, right = 0, x
    guess = (right + left)/2.0
    while abs (guess**2 - x) > epsilon:
        if guess **2 < x:
            left = guess
        else:
            right = guess
        guess = (right + left)/2.0
        iter += 1

    return guess, iter
```

```python
gc_sqrt (777777)[0]**2
```

- Newton-Ralphson method: x2 = C
  - f(x) = x^2 - C = 0
  - f'(x0) = f(x0)/(x0-x1)
  - x1 = x0 - (x0^2 - C) / 2x0

```python
# Newton-Raphson algorithm for sqrt(c)
epsilon = 0.001
max_iter = 100

def nr_sqrt2(c):
    xi = np.random.rand()
    i = 0
    while i < max_iter:
        diff = xi**2 - c
        if abs(diff) < epsilon:
            return xi, i
        else:
            xi = xi - diff / (2.0 * xi)
        i += 1

    return xi, i
```

# Operator overloading
- what is it?
  - gives an extended meaning beyond their predefined operational meaning.
  - (example) operator + is used to add two integers as well as join two strings and merge two lists.
  - It is achievable because ‘+’ operator is overloaded by int class and str class.

- Operator overloading
  - In Python, you can redefine the behavior of built-in operators (like +, -, *, /) for custom classes.
  - This allows you to create objects that can interact with operators in a way that aligns with their intended functionality.
- Mechanism:
  - Python achieves operator overloading through the concept of special methods or magic methods. These are methods with specific names that start and end with double underscores (__).
  - When you define these special methods within a class, Python associates them with the corresponding operators.
- Commonly Used Special Methods for Operator Overloading:
  - \_\_add__(self, other): Defines behavior for the + operator.
  - \_\_sub__(self, other): Defines behavior for the - operator.
  - \_\_mul__(self, other): Defines behavior for the * operator.
  - \_\_truediv__(self, other): Defines behavior for the / operator (true division).
  - \_\_floordiv__(self, other): Defines behavior for the // operator (floor division).
  - \_\_mod__(self, other): Defines behavior for the % operator (modulo).
  - \_\_pow__(self, other): Defines behavior for the ** operator (exponentiation).

```python
# Python program to show use of + operator for different purposes.

print(1 + 2)             # add two integers
print("Geeks "+"For")    # concatenate two strings

print(3 * 4)             # product of two numbers
print("Geeks "*4)        # repeat the String

(1).__add__(2), "Geeks ".__add__("For"), (3).__mul__(4), "Geeks ".__mul__(4)
```

```python
# Example: Complex Number Class with Operator Overloading:

class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __add__(self, other):

        # check the type of the object
        if isinstance(other, ComplexNumber):
            return ComplexNumber(self.real + other.real, self.imag + other.imag)
        else:
            return ComplexNumber(self.real + other, self.imag)

    def __repr__(self):
        return f"{self.real} + {self.imag}j"

    def __str__(self):
        return f"{self.real} + {self.imag}j"

# Create complex numbers
c1 = ComplexNumber(2, 3)
c2 = ComplexNumber(4, 1)
```

- generator: yield statement 를 가지고 있는 함수 (Iterator를 생성해 준다)
  - 호출할 때마다 한 개의 값을 리턴
  - yield가 호출되면 임시적으로 return이 호출되며, 한번 더 실행되면 실행되었던 이전 yield의 다음 코드가 실행

```python
# generator
def generator_():
    print("aa")
    yield "a"
    print("bb")
    yield "b"
    print("cc")
    yield "c"
gen = generator_()
gen
```

```python
next(gen) # aa, a
```

```python
next(gen) # bb, b
```

```python
next(gen) # cc, c
```

```python
# inheritance
import numpy as np
class Class1:

    def __init__(self, arr: list):
        self.arr1 = np.array(arr)
    def get_max(self):
        return(np.max(self.arr1))

class Class2 (Class1):

    def __init__(self, arr: list):
        super().__init__(arr)
        self.arr2 = arr
    def get_argmax(self):
        return(np.argmax(self.arr2))
```

```python
a1 = Class1([1, 4, 2, 3, 5, 6, 19])
a2 = Class2([5, 6, 9, 8, 3, 4, 11])
print(a1.get_max())
print(a2.get_argmax())

try:
    print(a2.get_max())
except:
    print("Error!")
```
